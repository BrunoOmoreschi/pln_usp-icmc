2 - Choose a corpus (you can use the one from classes) and using NLTK functions, do:
    a. Tokenize the entire corpus (words, numbers and punctuations)
    b. Verify how many tokens are in corpus
    c. Tokenize the corpus using only words
    d. Verify how many tokens are in corpus by word
    e. Verify the frequency of the words in corpus
    f. Verify which are the most 5, 10 and 15 most frequent words in corpus
    g. Extract the stopwords from NLTK
    h. Verify the frequency of tokens without stopwords in corpus

```python
#2.a
import nltk
corpusFile = open('corpus_teste.txt','r')
corpus = corpusFile.read()
nltkTokens = nltk.word_tokenize(corpus)
nltkTokens
```

```python
#2.b
len(nltkTokens)
```

```python
#2.c
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'[a-zA-Z]\w*')
wordTokens = tokenizer.tokenize(corpus)
wordTokens
```

```python
#2.d
len(wordTokens)
```

```python
#2.e
frequency = nltk.FreqDist(wordTokens)
frequency.most_common()
```

```python
#2.f
print(f"The 5 most used words in the text are: {frequency.most_common(5)}")
print(f"\nThe 10 most used words in the text are: {frequency.most_common(10)}")
print(f"\nThe 15 most used words in the text are: {frequency.most_common(15)}")
```

```python
#2.g
stopwords = nltk.corpus.stopwords.words('portuguese')
stopwords
```

```python
#2.h. Verify the frequency of tokens without stopwords in corpus

#We need to set all the words on lower case standard to count it right. If not the same word may be counted twice.

noStopWordTokens = []

for token in wordTokens:
    if token.lower() not in stopwords:
        noStopWordTokens.append(token.lower())

cleanFrequency = nltk.FreqDist(noStopWordTokens)
cleanFrequency.most_common()
```
